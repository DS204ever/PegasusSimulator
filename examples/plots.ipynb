```python
import numpy as np
import matplotlib.pyplot as plt

# Load the position error
statistics = np.load('results/statistics_1.npz')
statistics2 = np.load('results/statistics_2.npz')

time1 = statistics["time"]

# Load the statistics for vehicle 1
p1 = statistics["p"]
ep1 = statistics["ep"]
ev1 = statistics["ev"]
er1 = statistics["er"]
ew1 = statistics["ew"]

# Load the statistics for vehicle 2
p2 = statistics2["p"][1:]
ep2 = statistics2["ep"][1:]
ev2 = statistics2["ev"][1:]
er2 = statistics2["er"][1:]
ew2 = statistics2["ew"][1:]
```

# Vehicles 1 and 2

```python
# Trajectory executed by the 2 vehicles
ax = plt.figure().add_subplot(projection='3d')
ax.plot(p1[:,0], p1[:,1], p1[:,2], label='Vehicle 1')
ax.plot(p2[:,0], p2[:,1], p2[:,2], label='Vehicle 2', color='red')
ax.view_init(elev=22, azim=-3)
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.savefig("results/trajectory.svg", format = 'svg', dpi=300)
ax.legend()
```

```python
# Compyte the distance between the 2 vehicles
plt.figure()
plt.title("Distance between the two vehicles")
plt.plot(time1, np.linalg.norm(p1-p2, axis=1))
plt.ylabel("Distance [m]")
plt.grid()
plt.savefig("results/distance_between_vehicle.svg", format = 'svg', dpi=300)
```

```python
plt.figure()
# Error in the X-axis
plt.subplot(3,1,1)
plt.plot(time1, ep1[:,0])
plt.plot(time1, ep2[:,0], color='red')
plt.title('Position Error')
plt.ylabel("X [m]")
plt.ylim([-0.35, 0.35])
plt.legend(["Vehicle 1", "Vehicle2"])
plt.grid()

# Error in the Y-axis
plt.subplot(3,1,2)
plt.plot(time1, ep1[:,1])
plt.plot(time1, ep2[:,1], color='red')
plt.ylabel("Y [m]")
plt.ylim([-0.60, 0.60])
plt.grid()

# Error in the Z-axis
plt.subplot(3,1,3)
plt.plot(time1, ep1[:,2])
plt.plot(time1, ep2[:,2], color='red')
plt.ylabel("Z [m]")
plt.ylim([-0.35, 0.35])
plt.grid()

plt.savefig("results/tracking_error.svg", format = 'svg', dpi=300)
plt.show()
```

# Vehicle 1

```python
# Plot the position error
plt.figure()
plt.subplot(4,1,1)
plt.plot(time1, ep1[:,0])
plt.plot(time1, ep1[:,1])
plt.plot(time1, ep1[:,2])
plt.ylabel("Position error [m]")
plt.xlabel("Time [s]")
plt.title('Position error [m]')
plt.legend(["x-axis", "y-axis", "z-axis"])
plt.grid()
plt.show()

# Plot the velocity error
plt.figure()
plt.subplot(4,1,2)
plt.plot(time1, ev1[:,0])
plt.plot(time1, ev1[:,1])
plt.plot(time1, ev1[:,2])
plt.ylabel("Velocity error [m/s]")
plt.xlabel("Time [s]")
plt.title('Velocity error [m/s]')
plt.legend(["x-axis", "y-axis", "z-axis"])
plt.grid()
plt.show()

# Plot the attitude error
plt.figure()
plt.subplot(4,1,3)
plt.plot(time1, er1[:,0])
plt.plot(time1, er1[:,1])
plt.plot(time1, er1[:,2])
plt.ylabel("Attitude error [rad]")
plt.xlabel("Time [s]")
plt.title('Attitude error [rad]')
plt.legend(["roll", "pitch", "yaw"])
plt.grid()
plt.show()

# Plot the angular velocity error
plt.figure()
plt.subplot(4,1,4)
plt.plot(time1, ew1[:,0])
plt.plot(time1, ew1[:,1])
plt.plot(time1, ew1[:,2])
plt.ylabel("Attitude-rate error [rad/s]")
plt.xlabel("Time [s]")
plt.title('Attitude-rate error [rad/s]')
plt.legend(["roll-rate", "pitch-rate", "yaw-rate"])
plt.grid()
plt.show()
```
