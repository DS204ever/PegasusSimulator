```python
import numpy as np
import matplotlib.pyplot as plt

# Load the position error
statistics = np.load('results/single_statistics.npz')

time = statistics["time"]
ep = statistics["ep"]
ev = statistics["ev"]
er = statistics["er"]
ew = statistics["ew"]
```

```python
# Plot the position error
plt.figure()
plt.plot(time, ep[:,0])
plt.plot(time, ep[:,1])
plt.plot(time, ep[:,2])
plt.ylabel("Position error [m]")
plt.xlabel("Time [s]")
plt.title('Position error [m]')
plt.legend(["x-axis", "y-axis", "z-axis"])
plt.grid()
plt.show()
```

```python
# Plot the velocity error
plt.figure()
plt.plot(time, ev[:,0])
plt.plot(time, ev[:,1])
plt.plot(time, ev[:,2])
plt.ylabel("Velocity error [m/s]")
plt.xlabel("Time [s]")
plt.title('Velocity error [m/s]')
plt.legend(["x-axis", "y-axis", "z-axis"])
plt.grid()
plt.show()
```

```python
# Plot the attitude error
plt.figure()
plt.plot(time, er[:,0])
plt.plot(time, er[:,1])
plt.plot(time, er[:,2])
plt.ylabel("Attitude error [rad]")
plt.xlabel("Time [s]")
plt.title('Attitude error [rad]')
plt.legend(["roll", "pitch", "yaw"])
plt.grid()
plt.show()
```

```python
# Plot the angular velocity error
plt.figure()
plt.plot(time, ew[:,0])
plt.plot(time, ew[:,1])
plt.plot(time, ew[:,2])
plt.ylabel("Attitude-rate error [rad/s]")
plt.xlabel("Time [s]")
plt.title('Attitude-rate error [rad/s]')
plt.legend(["roll-rate", "pitch-rate", "yaw-rate"])
plt.grid()
plt.show()
```
